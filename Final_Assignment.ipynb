<h3>My code will print n<sup>th</sup> element of Fibonacci series</h3>

```python
def fib(n):
    if n == 0:
        return 0
    if n == 1:
        return 1
    ele = fib(n-1) + fib(n-2)
    return(ele)
n = 7
print(fib(n))
```
